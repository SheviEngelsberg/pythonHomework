```python
studyHours = [2,3,4,5,6,5,4,3,2,1]
marks = [100,98,98,87,98,78,77,88,99,100]
colors = np.random.rand(10) 
sizes = 1000 * np.random.rand(10)
plt.scatter(studyHours, marks, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()  # הצגת סרגל הצבעים
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
ages = [22, 45, 30, 35, 50, 41, 28, 33, 55, 65, 29, 42, 48, 31]

plt.hist(ages, bins=[20, 30, 40, 50, 60, 70], edgecolor='black')

plt.title('Age Distribution')
plt.xlabel('Ages')
plt.ylabel('Number of People')

```

```python
years = range(2010, 2021)
consumption_oil = [100, 105, 110, 120, 125, 130, 135, 140, 145, 150, 155]
consumption_gas = [80, 82, 85, 88, 90, 95, 100, 105, 110, 115, 120]
consumption_renewable1 = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]
consumption_renewable2 = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]

fig, axs = plt.subplots(2,2)  # יוצר מטריצה של גרפים בגודל 2x2
axs[0, 0].plot(years, consumption_oil, 'tab:blue')
axs[0, 0].set_title('1')
axs[0, 1].plot(years, consumption_gas, 'tab:red')
axs[0, 1].set_title('2')
axs[1, 0].scatter(years,consumption_renewable1, c='yellow')
axs[1, 0].set_title('3')
axs[1, 1].scatter(years,consumption_renewable2, c='green')
axs[1, 1].set_title('4')

# התאמת הגרפים למראה נקי יותר
for ax in axs.flat:
    ax.label_outer()
```
